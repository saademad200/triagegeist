### 3️ What is Epilepsy

**Epilepsy** is a **chronic neurological disorder** characterized by **recurrent seizures** caused by abnormal electrical activity in the brain.

Seizures may cause:

- loss of consciousness  
- uncontrolled muscle movements  
- confusion or disorientation  
- temporary behavioral changes  

Triggers for seizures can include:

- stress  
- sleep deprivation  
- infections  
- medication changes  
- metabolic disturbances  

---

### 4️ Why This Feature Is Important in Emergency Triage

Patients with epilepsy may present to the emergency department due to:

- **active seizures**
- **status epilepticus** (continuous seizures)
- **post-seizure confusion**
- **injuries caused during seizures**

Status epilepticus is a **life-threatening neurological emergency** that requires **immediate medical treatment**.

Additionally, seizures may also indicate **serious underlying problems** such as:

- brain infection  
- head injury  
- metabolic imbalance  

---

### 5️ Clinical Importance During Triage

Emergency clinicians consider epilepsy history when evaluating symptoms such as:

- **recent seizure activity**
- **post-ictal state (recovery phase after a seizure)**
- **neurological instability**

These cases may require **urgent neurological evaluation**.

### 3️ What is Coagulopathy

**Coagulopathy** refers to a group of conditions where the **blood clotting process is abnormal**.

Normally, when a person is injured, the body forms a **blood clot** to stop bleeding.  
In coagulopathy, this process may be **impaired or excessive**.

Types of clotting disorders include:

- **Bleeding disorders** (blood does not clot properly)
- **Hypercoagulable states** (blood clots too easily)

Examples of conditions that cause coagulopathy:

- hemophilia  
- liver disease  
- vitamin K deficiency  
- anticoagulant medication use (e.g., warfarin)  
- disseminated intravascular coagulation (DIC)

---

### 4️ Why This Feature Is Important in Emergency Triage

Coagulopathy is clinically important in emergency medicine because abnormal clotting can lead to **serious complications**, including:

- **severe bleeding**
- **internal hemorrhage**
- **stroke**
- **dangerous blood clots**
- **complications during trauma**

Patients with clotting disorders are at higher risk during situations such as:

- injuries  
- surgeries  
- infections  
- major trauma

Even minor injuries can result in **excessive bleeding**.

### 3️ What is Coronary Artery Disease

**Coronary Artery Disease (CAD)** occurs when the **coronary arteries that supply oxygen-rich blood to the heart become narrowed or blocked** due to plaque buildup.

Common symptoms include:

- chest pain (angina)  
- shortness of breath  
- fatigue  
- discomfort in the arms, neck, or jaw  

In severe cases, CAD can lead to **heart attacks**.

---

### 3️ What is a Stroke

A **stroke** occurs when the **blood supply to part of the brain is interrupted or reduced**, preventing brain tissue from receiving oxygen and nutrients.

There are two major types of stroke:

- **Ischemic stroke** – caused by a blood clot blocking a brain artery  
- **Hemorrhagic stroke** – caused by bleeding in the brain

Stroke can cause damage to brain cells within minutes.

Common symptoms include:

- sudden weakness on one side of the body  
- difficulty speaking or understanding speech  
- vision problems  
- severe headache  
- loss of balance or coordination  

---

### 4️ Why This Feature Is Important in Emergency Triage

Patients with a prior stroke are at **significantly higher risk of future neurological events**, including:

- **recurrent stroke**
- **transient ischemic attacks (TIAs)**
- **neurological deterioration**
- **vascular complications**

Stroke survivors often have **underlying vascular disease**, making them more vulnerable to further brain injury.

### 3️ What is Peripheral Vascular Disease

**Peripheral Vascular Disease (PVD)** refers to conditions where **blood vessels outside the heart and brain become narrowed, blocked, or damaged**.

It commonly affects the **arteries supplying blood to the legs and arms**.

The most common cause is **atherosclerosis**, where fatty plaques build up inside arteries and restrict blood flow.

Common symptoms include:

- leg pain while walking (claudication)  
- numbness or weakness in limbs  
- coldness in lower legs or feet  
- slow-healing wounds on the legs or feet  

---

### 4️ Why This Feature Is Important in Emergency Triage

Peripheral vascular disease is important in emergency medicine because it often indicates **widespread vascular disease throughout the body**.

Patients with PVD have increased risk of:

- **heart attack**
- **stroke**
- **blood clots**
- **limb ischemia (reduced blood supply to limbs)**

Severe cases can lead to **critical limb ischemia**, where tissues begin to die due to lack of blood flow.

---

### 5️ Clinical Importance During Triage

Emergency clinicians consider PVD history when evaluating symptoms such as:

- leg pain or numbness  
- cold or pale limbs  
- wounds that do not heal  
- sudden loss of blood flow to a limb  

Example scenario:

Patient arrives with:

- severe leg pain  
- cold and pale foot  
- reduced pulse in the limb  

If **hx_peripheral_vascular_disease = 1**, clinicians may suspect:

- **acute limb ischemia**
- **vascular blockage**
- **circulatory complications**

These conditions require **urgent vascular assessment and treatment**.
